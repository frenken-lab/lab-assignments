## Part 3 — Compare pip vs uv speed

Run both installs and capture the wall-clock time. `!time ...` works in a `bash` cell.


```python
%%bash
python -m venv pip-test
time pip-test/bin/pip install --quiet requests flask pandas numpy
rm -rf pip-test
```

```python
%%bash
# Install uv first if not already on PATH:
#   curl -LsSf https://astral.sh/uv/install.sh | sh
uv venv uv-test
time uv pip install --quiet --python uv-test/bin/python requests flask pandas numpy
rm -rf uv-test
```

## Part 4 — Requirements files

Generate a `pip freeze` and reproduce the environment in a fresh venv.


```python
!uv venv req-test
!uv pip install --python req-test/bin/python requests pandas matplotlib
!req-test/bin/pip freeze > requirements.txt
!cat requirements.txt
```

```python
!uv venv req-verify
!uv pip install --python req-verify/bin/python -r requirements.txt
!req-verify/bin/python -c "import requests, pandas, matplotlib; print('All packages installed successfully')"
```
